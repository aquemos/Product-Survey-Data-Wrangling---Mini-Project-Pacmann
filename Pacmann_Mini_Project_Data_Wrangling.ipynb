Based on the example data above, answer D in the data is accompanied by the value of the option, namely "D. Tidak memilih semua product". Meanwhile, options A, B, and C do not include the value of the option. Since the format is inconsistent, we will change the format from option D

##Reformat option D format

```python
# Function to reformat choice D by removing a specific text pattern.
# If the input 'choice' contains the text pattern ". Tidak memilih semua product",
# this function removes that pattern and returns the modified choice. Otherwise,
# it returns the original 'choice' unchanged.
# example: Input: 'A, D. Tidak memilih semua product' -> Output: 'A, D'

def reformat_choice_D(choice):
  text_to_replace = ". Tidak memilih semua product"
  if choice.find(text_to_replace) != -1:
    return choice.replace(text_to_replace, "")
  else:
    return choice
```

```python
columns_to_process = [col for col in data.columns if col not in [data.columns[0], data.columns[1]]]
columns_to_process
```

```python
# Create a list of columns to process by excluding the first two columns from the 'data' DataFrame.
# These first two columns are TimeStamp and user phone number.
data[columns_to_process] = data[columns_to_process].applymap(reformat_choice_D)

data
```

```python
# Create a list named that contains ten dictionaries above.
questions_option = [question_1,
                    question_2,
                    question_3,
                    question_4,
                    question_5,
                    question_6,
                    question_7,
                    question_8,
                    question_9,
                    question_10]
```

```python
# Function to extract and process survey choices.
# The function takes three parameters:
# - 'cleaned_data': A DataFrame to store cleaned and processed data.
# - 'phone_num': The user's phone number associated with the choice.
# - 'choice': The user's survey choice (a comma-separated string).
# - 'question': A dictionary containing options and their attributes for the survey question.

def extract_choice(cleaned_data, phone_num, choice, question):
  # Split the 'choice' string into a list of individual choices.
  choice_list = choice.split(", ")


  # Check if there is more than one choice and if 'D' is in the choices.
  # If so, the answer is ignored.
  # Because the answer is invalid (example: user choice A and D, this is invalid choice)
  if len(choice_list) > 1 and 'D' in choice_list:
    pass

  else:
    # Iterate through the question dictionary to identify selected choices.
    for key, value in question.items():
      if key in choice_list:
        choice = 1 # set choice = 1, if the option in the question is selected in choice list
      else:
        choice = 0 # set choice = 0, if the option in the question is not selected in choice list

      # Assign the 'phone_num' and 'choice' values to the dictionary.
      value['user_phone'] = phone_num
      value['choice'] = choice

      # Create a DataFrame from the dictionary and concatenate it with 'cleaned_data'.
      extracted_choice = pd.DataFrame(value, index=[0])
      cleaned_data = pd.concat([cleaned_data, extracted_choice], ignore_index=True)

  return cleaned_data # Return the updated 'cleaned_data' DataFrame.
```

```python
for index, row in data.iterrows(): # Iterate through each row in the 'data' DataFrame
  phone_num = row[data.columns[1]] # Extract the user's phone number from the current row.

  for num in range(10): # Iterate through 10 questions of survey
    # Get the user's choice from each columns of 'data' DataFrame
    # Num + 2, because the choice data in the 'data' dataframe starts from the 2nd column index
    choice = row[data.columns[num + 2]]

    # Get question attributes from 'questions_option' using 'num' as an index.
    question_attributes = questions_option[num]

    # Call the 'extract_choice' function to process and extract the choice information,
    # and update the 'cleaned_data' DataFrame accordingly.
    cleaned_data = extract_choice(cleaned_data, phone_num, choice, question_attributes)
```
